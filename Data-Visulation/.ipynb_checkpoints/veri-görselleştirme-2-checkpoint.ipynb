```python
1- barplot
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x="sex",y ="survived",data=df);  # noktalı virgüle dikkat
```

```python
sns.barplot(x = "survived", y = "sex", data = df);
```

```python
sns.barplot(x = "class", y = "survived", data = df);
```

```python
sns.barplot(x = "sex", y = "survived", hue = "class", data = df);
```

```python
sns.barplot(x = "sex", y = "survived", data = df);
```

```python
tips = sns.load_dataset("tips")
```

```python
df = tips.copy()
```

```python
df.head()
```

```python
df.groupby("day").mean()
```

```python
df.isnull().sum()  # eksik verileri kontrol ettik.
```

```python
df["sex"].value_counts()  # cinsiyetlerin sayısı
```

```python
df["day"].value_counts()
```

```python
sns.barplot(x = "sex", y = "total_bill", data = df);
# baylar bayanlardan daha fazla harcama yapmışlar
```

```python
sns.barplot(x = "sex", y = "total_bill", hue = "time", data = df);
```

```python
sns.barplot(x = "sex", y = "tip", data = df);
```

```python
sns.barplot(x = "sex" , y = "tip", hue = "smoker", data = df);
```

```python
sns.barplot(x = "day", y = "total_bill", hue = "time", data =df );
```

```python
# 2- boxplot grafiği 
```

```python
sns.boxplot(x = "sex",  y = "total_bill", data = df);
```

```python
sns.boxplot(x = "sex", y = "total_bill", hue = "time", data =df );
```

```python
# 3-vionplot
```

```python
sns.catplot(x = "sex", y = "tip", kind = "violin", data = df);
```

```python
sns.catplot(x = "sex", y = "total_bill", hue = "day", kind = "swarm",data = df);
```

```python
# scatterplot grafiği
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```
