```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head(3)
```

```python
# fonkisyon uygulama 
# aggregate fonksiyonu
df.dtypes
```

```python
df_number = df.select_dtypes(include=["float64","int64"])  # belli veri tipini se√ßtik. 
df_number.head()
```

```python
df_number.aggregate(["min","max"])  # b√ºt√ºn s√ºt√ºnlarƒ±n i√ßinde bulunan max ve min deƒüerler
```

```python
df["age"].min()
```

```python
df_number.aggregate([np.median])  # medyan deƒüerlerini ekran yazdƒ±rdƒ±k
```

```python
df_number.aggregate(["mean","median"])
```

```python
df_number.aggregate(["mean",np.median])
```

```python
df_number.aggregate({"survived":"mean","pclass":min,"age":max})  # s√∂zl√ºk yapƒ±sƒ± olarak da verileri √ßekebiliriz
```

```python
def f(x):
    return x.mean()-x.std()
```

```python
df_number.aggregate(f)  # 27 ve 28. h√ºcreler aynƒ±
```

```python
f(df_number)   # 27 ve 28. h√ºcreler aynƒ±
```

```python
df_number.aggregate(["std","mean",f])
```

```python
fonksiyon = ["max","mean","std"]
```

```python
df_number.aggregate(fonksiyon)
```
